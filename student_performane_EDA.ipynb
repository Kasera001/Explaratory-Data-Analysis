# Exploratory Data Analysis

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
#
```

```python
# reading the data set and finding its data types

df = pd.read_csv("/content/student_data.csv")
df.info()
```

```python
print(df.head()) # prints the first 5 elements
print(df.tail()) # prints the last 5 rows
df.describe()
```

```python
# Checking for null values and duplicates
print("Missing values per column:")
display(df.isnull().sum())
df.duplicated().sum()
df.dropna()

print("\nNumber of duplicate rows:", df.duplicated().sum())
```

```python
plt.figure(figsize=(10, 6))
sns.countplot(x='studytime', hue='failures', data=df)
plt.xlabel("Study Time (1: <2h, 2: 2-5h, 3: 5-10h, 4: >10h)")
plt.ylabel("Number of Students")
plt.title("Distribution of Failures by Study Time")
plt.legend(title='Failures')
plt.show()

# you can clearly see that students who spend more time studying
# (categories 3 and 4) tend
# to have fewer failures compared to those in category 1.
```

```python
plt.figure(figsize=(8, 6))
sns.barplot(x='Pstatus', y='G3', data=df, palette='Set2')
plt.title('Comparison of Final Grades (G3) by Parent Status')
plt.xlabel('Parental Status (T: Together, A: Apart)')
plt.ylabel('Final Grade (G3)')
plt.show()
# The analysis shows that students whose parents live apart ('A')
# have a slightly higher mean final grade (~11.20)
# compared to those whose parents live together ('T', ~10.32).

# Calculate mean grades for numerical comparison
mean_grades = df.groupby('Pstatus')['G3'].mean().reset_index()
print("Mean Final Grade by Parent Status:")
display(mean_grades)
```

```python
plt.figure(figsize=(10, 6))
sns.regplot(x='absences', y='G3', data=df, scatter_kws={'alpha':0.5}, line_kws={'color':'red'})
plt.title('Relationship between Absences and Final Grade (G3)')
plt.xlabel('Number of Absences')
plt.ylabel('Final Grade (G3)')
plt.grid(True, linestyle='--', alpha=0.6)
plt.show()

# Calculate correlation coefficient
correlation = df['absences'].corr(df['G3'])
print(f"Correlation coefficient between absences and G3: {correlation:.3f}")
```

```python
plt.figure(figsize=(8, 6))
sns.boxplot(x='school', y='G3', data=df, palette='Pastel1')
plt.title('Distribution of Final Grades (G3) by School')
plt.xlabel('School')
plt.ylabel('Final Grade (G3)')
plt.grid(axis='y', linestyle='--', alpha=0.5)
plt.show()
#The boxplot and statistics show that students at school 'GP' (Gabriel Pereira)
#have a slightly higher mean final grade (10.49) compared
#to school 'MS' (Mousinho da Silveira, 9.85).
#While the median for GP is 11 and for MS is 10,
#both schools show a similar range of scores, including some students scoring 0 and
#others reaching high marks near 20.

# Calculate summary statistics by school
school_stats = df.groupby('school')['G3'].describe()
print("Summary Statistics for G3 by School:")
display(school_stats)
```

```python
plt.figure(figsize=(8, 6))
sns.barplot(x='internet', y='G3', data=df, palette='viridis')
plt.title('Average Final Grade (G3) by Internet Access')
plt.xlabel('Has Internet Access (yes/no)')
plt.ylabel('Average Final Grade (G3)')
plt.show()
#The analysis shows that students with internet access ('yes') have
#a higher average final grade (10.62) compared to those without it (9.41).
#While there is a noticeable difference in the mean,
#it's also worth noting that the 'yes' group
#is much larger (329 students) than the 'no' group (66 students).


# Calculate statistics for internet access
internet_stats = df.groupby('internet')['G3'].agg(['count', 'mean', 'std']).reset_index()
print("Summary Statistics for G3 by Internet Access:")
display(internet_stats)
```

```python
plt.figure(figsize=(8, 6))
sns.barplot(x='schoolsup', y='G3', data=df, palette='Set1')
plt.title('Average Final Grade (G3) by School Educational Support')
plt.xlabel('Extra Educational Support (yes/no)')
plt.ylabel('Average Final Grade (G3)')
plt.show()
# the analysis show that students with no educational  support have a higher
#avarage final grade compared to those with financial support
# Calculate statistics for school educational support
schoolsup_stats = df.groupby('schoolsup')['G3'].agg(['count', 'mean', 'std']).reset_index()
print("Summary Statistics for G3 by School Educational Support:")
display(schoolsup_stats)
```

```python
plt.figure(figsize=(8, 6))
sns.barplot(x='famsize', y='G3', data=df, palette='magma')
plt.title('Average Final Grade (G3) by Family Size')
plt.xlabel('Family Size (LE3: <=3, GT3: >3)')
plt.ylabel('Average Final Grade (G3)')
plt.show()
#he analysis of family size shows that students from smaller families (LE3: 3 or fewer members) have a slightly higher
# average final grade of 11.00 compared to 10.18 for those from larger familie
# Calculate statistics for family size
famsize_stats = df.groupby('famsize')['G3'].agg(['count', 'mean', 'std']).reset_index()
print("Summary Statistics for G3 by Family Size:")
display(famsize_stats)
```

```python
plt.figure(figsize=(8, 6))
sns.barplot(x='address', y='G3', data=df, palette='coolwarm')
plt.title('Average Final Grade (G3) by Address Type')
plt.xlabel('Address (U: Urban, R: Rural)')
plt.ylabel('Average Final Grade (G3)')
plt.show()

# Calculate statistics for address type
address_stats = df.groupby('address')['G3'].agg(['count', 'mean', 'std']).reset_index()
print("Summary Statistics for G3 by Address Type:")
display(address_stats)
```
